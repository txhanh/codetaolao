```python
def change_date_format(list,index,field):
    slice = list[index][field].split("-")
    if slice[1] == "JAN": 
        slice[1] = "01"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
        # print(slice)
    if slice[1] == "FEB": 
        slice[1] = "02"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "MAR": 
        slice[1] = "03"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "APR": 
        slice[1] = "04"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"      
        list[index][field] = slice  
    if slice[1] == "MAY": 
        slice[1] = "05"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "JUN": 
        slice[1] = "06"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "JUL": 
        slice[1] = "07"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "AUG": 
        slice[1] = "08"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "SEP": 
        slice[1] = "09"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "OCT": 
        slice[1] = "10"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "NOV": 
        slice[1] = "11"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "DEC": 
        slice[1] = "12"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
```

```python
for index in range(len(sourceList)):
    change_date_format(sourceList,index,'create_date')
    change_date_format(sourceList,index,'insert_time')
```
