```python
def change_date_format(list,index,field):
    slice = list[index][field].split("-")
    if slice[1] == "JAN": 
        slice[1] = "01"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
        # print(slice)
    if slice[1] == "FEB": 
        slice[1] = "02"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "MAR": 
        slice[1] = "03"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "APR": 
        slice[1] = "04"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"      
        list[index][field] = slice  
    if slice[1] == "MAY": 
        slice[1] = "05"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "JUN": 
        slice[1] = "06"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "JUL": 
        slice[1] = "07"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "AUG": 
        slice[1] = "08"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "SEP": 
        slice[1] = "09"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "OCT": 
        slice[1] = "10"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "NOV": 
        slice[1] = "11"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
    if slice[1] == "DEC": 
        slice[1] = "12"
        slice = f"20{slice[2]}-{slice[1]}-{slice[0]}"
        list[index][field] = slice
```

```python
for index in range(len(sourceList)):
    change_date_format(sourceList,index,'create_date')
    change_date_format(sourceList,index,'insert_time')
```

```python
 # initializing list
tenant_id_list = ['CLMX','ABBOTT','VIFON','VINAMILK']
staff_title_list = ['Nhan vien ban hang','Giam sat','ASM','RSM', 'Nhan vien ban hang','Nhan vien ban hang','Nhan vien ban hang','Nhan vien ban hang']
os_version_list = ['1.1','1.2','1.3','1.4']
app_version_list = ['1.8','1.9', '2.0']
speed_list = ['WEAK','STRONG','EXTREME'] 
# # printing original list
# print("Original list is : " + str(tenant_id_list))
 
# # using random.choice() to
# # get a random number
# random_num = random.choice(tenant_id_list)
 
# # printing random number
# print("Random selected number is : " + str(random_num))
```

```python
sinkList = []
# for index in range(0, len(sourceList)):
#     for key in sourceList[index]
for document in sourceList:
    sinkList.append(
        {
            'tenant_id': random.choice(tenant_id_list),
            'staff_id': str(document['staff_id']),
            'log_date': f'new Date("{document["create_date"]}")',
            'last_lat': document['lat'],
            'last_lng': document['lng'],
            'logs': [
                {
                    'staff_title': random.choice(staff_title_list),
                    'distribution_unit_id': str(document['shop_id']),
                    'lat': document['lat'],
                    'lng': document['lng'],
                    'distance': document['distance'],
                    'accuracy': document['accuracy'],
                    'battery': document['pin'],
                    'network_type': document['network_type'],
                    'speed': random.choice(speed_list),
                    'device_id': ''.join(random.choices(string.ascii_uppercase + string.ascii_lowercase, k=10)),
                    'os_version': random.choice(os_version_list),
                    'app_version': random.choice(app_version_list),
                    'create_date': f'new Date("{document["insert_time"]}")'
                }
            ]
        }
    )
    
```
